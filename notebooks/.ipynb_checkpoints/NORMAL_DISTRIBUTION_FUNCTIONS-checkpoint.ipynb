# Normal Distribution -- A.K.A. Gaussian Distribution

## Probability Distribution Function

\begin{equation*}
f(x \ |\ \mu, \sigma^2) = \frac{1}{\sqrt{2 \pi \sigma^2 }} \  \exp \left(-\frac{(x-\mu)^2}{2 \sigma^2}\right)
\end{equation*}


$\mu \ \ $ is mean or expectation of the distribution (and also its median and mode).

$\sigma \ \ $ is the standard deviation.

$\sigma^2 \ \ $ is the variance.



## Cumulative Distribution Function

\begin{equation*}
CDF = \frac{1}{2} \left[ 1 + \text{erf} \left( \frac{x-\mu}{\sigma \sqrt{2}} \right) \right]
\end{equation*}


$\mu \ \ $ is mean or expectation of the distribution (and also its median and mode).

$\sigma \ \ $ is the standard deviation.

$\sigma^2 \ \ $ is the variance.

erf    is the error function.

## Quantile 

\begin{equation*}
\mu \ + \ \sigma \sqrt{2} \ \text{erf}^{-1}(2F-1) 
\end{equation*}


To find the probability (P) confidence interval centered about the mean for a normal distribution in units of $\sigma$, 
\begin{equation*}
n = \sqrt{2} \text{erf}^{-1}(P),
\end{equation*}
where 

$P \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ X_P \ \ \ $

$0.800 \ \ \ \ \ \ \ \ \ 1.28155 \ \sigma \ \ \ $

$0.900 \ \ \ \ \ \ \ \ \ 1.64485 \ \sigma \ \ \ $

$0.950 \ \ \ \ \ \ \ \ \ 1.95996 \ \sigma \ \ \ $

$0.990 \ \ \ \ \ \ \ \ \ 2.57583 \ \sigma \ \ \ $

$0.995 \ \ \ \ \ \ \ \ \ 2.80703 \ \sigma \ \ \ $

$0.999 \ \ \ \ \ \ \ \ \ 3.29053 \ \sigma \ \ \ $

```python
def n(P):
    return numpy.sqrt(2) * erfinv(P)

n(.8), n(.9), n(.95), n(.990), n(.995), n(.999)
```

```python
class Normal(object):
    
    def __init__(self):
        ""
    def mean(self, Mu, Sigma=None):
        return Mu
    
    def median(self, Mu, Sigma=None):
        return Mu
    
    def mode(self, Mu, Sigma=None):
        return Mu
    
    def var(self, Mu=None, Sigma):
        return Sigma**2
        
    def quantile(self, Probs, Mu=None, Sigma=None): # all None's have to be last
        probs = [Probs]
        probs = numpy.asanyarray(probs)
        if not Mu:
            Mu    = 0
        if not Sigma:
            Sigma = 1.0
        quant = []
        for i,prob in enumerate(probs):
            quant.append(Mu + Sigma * numpy.sqrt(2) * erfinv(2 * prob - 1))
        return quant[0]
    
    def pdf(self, X, Mu, Sigma):
        return (1./(Sigma * numpy.sqrt(2 * numpy.pi))) * numpy.exp( - ((X - mu)**2) / (2 * Sigma**2))
    
Norm = Normal()
```
