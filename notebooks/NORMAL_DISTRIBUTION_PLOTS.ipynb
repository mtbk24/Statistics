# Normal Distribution -- A.K.A. Gaussian Distribution

## Probability Distribution Function

\begin{equation*}
f(x \ |\ \mu, \sigma^2) = \frac{1}{\sqrt{2 \pi \sigma^2 }} \  \exp \left(-\frac{(x-\mu)^2}{2 \sigma^2}\right)
\end{equation*}


$\mu \ \ $ is mean or expectation of the distribution (and also its median and mode).

$\sigma \ \ $ is the standard deviation.

$\sigma^2 \ \ $ is the variance.



## Cumulative Distribution Function

\begin{equation*}
CDF = \frac{1}{2} \left[ 1 + \text{erf} \left( \frac{x-\mu}{\sigma \sqrt{2}} \right) \right]
\end{equation*}


$\mu \ \ $ is mean or expectation of the distribution (and also its median and mode).

$\sigma \ \ $ is the standard deviation.

$\sigma^2 \ \ $ is the variance.

erf    is the error function.

## Quantile 

\begin{equation*}
\mu \ + \ \sigma \sqrt{2} \ \text{erf}^{-1}(2F-1) 
\end{equation*}


To find the probability (P) confidence interval centered about the mean for a normal distribution in units of $\sigma$, 
\begin{equation*}
n = \sqrt{2} \text{erf}^{-1}(P),
\end{equation*}
where 

$P \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ \ X_P \ \ \ $

$0.800 \ \ \ \ \ \ \ \ \ 1.28155 \ \sigma \ \ \ $

$0.900 \ \ \ \ \ \ \ \ \ 1.64485 \ \sigma \ \ \ $

$0.950 \ \ \ \ \ \ \ \ \ 1.95996 \ \sigma \ \ \ $

$0.990 \ \ \ \ \ \ \ \ \ 2.57583 \ \sigma \ \ \ $

$0.995 \ \ \ \ \ \ \ \ \ 2.80703 \ \sigma \ \ \ $

$0.999 \ \ \ \ \ \ \ \ \ 3.29053 \ \sigma \ \ \ $

## Probability Density Function (PDF)

Matching the Wikipedia page:
https://en.wikipedia.org/wiki/Normal_distribution

```python
plt.figure(figsize=(11,7))
x = numpy.linspace(-5, 5, 1000)
# loc:    mu or mean
# scale:  sigma or std dev, sqrt(variance)
plt.plot(x, norm.pdf(x, loc=0, scale = numpy.sqrt(0.2)), 'b', label='$\mu=0, \ \sigma^2 = 0.2$')
plt.plot(x, norm.pdf(x, loc=0, scale = numpy.sqrt(1.0)), 'r', label='$\mu=0, \ \sigma^2 = 1.0$')
plt.plot(x, norm.pdf(x, loc=0, scale = numpy.sqrt(5.0)), 'y', label='$\mu=0, \ \sigma^2 = 5.0$')
plt.plot(x, norm.pdf(x, loc=-2, scale = numpy.sqrt(0.5)), 'g', label='$\mu=-2, \ \sigma^2 = 0.5$')
plt.xlim([-5,5])
plt.legend(loc='best')
plt.title("Probability Distribtion Function \n Red curve is that standard normal distribution")
```

## Cumulative Distribution Function (CDF)

```python
plt.figure(figsize=(11,7))
x = numpy.linspace(-5, 5, 1000)
# loc:  mu or mean
# scale: sigma, scale, or std dev, sqrt(variance)
plt.plot(x, norm.cdf(x, loc=0, scale = numpy.sqrt(0.2)), 'b', label='$\mu=0, \ \sigma^2 = 0.2$')
plt.plot(x, norm.cdf(x, loc=0, scale = numpy.sqrt(1.0)), 'r', label='$\mu=0, \ \sigma^2 = 1.0$')
plt.plot(x, norm.cdf(x, loc=0, scale = numpy.sqrt(5.0)), 'y', label='$\mu=0, \ \sigma^2 = 5.0$')
plt.plot(x, norm.cdf(x, loc=-2, scale = numpy.sqrt(0.5)), 'g', label='$\mu=-2, \ \sigma^2 = 0.5$')
plt.xlim([-5,5])
plt.legend(loc='best')
plt.title("Cumulative Distribtion Function \n Red curve is that standard normal distribution")
```

```python
plt.figure(figsize=(11,7))
x = numpy.linspace(-8, 5, 1000)
# loc:  mu or mean
# scale: sigma, scale, or std dev, sqrt(variance)
plt.plot(x, norm.logcdf(x, loc=0, scale = numpy.sqrt(0.2)), 'b', label='$\mu=0, \ \sigma^2 = 0.2$')
plt.plot(x, norm.logcdf(x, loc=0, scale = numpy.sqrt(1.0)), 'r', label='$\mu=0, \ \sigma^2 = 1.0$')
plt.plot(x, norm.logcdf(x, loc=0, scale = numpy.sqrt(5.0)), 'y', label='$\mu=0, \ \sigma^2 = 5.0$')
plt.plot(x, norm.logcdf(x, loc=-2, scale = numpy.sqrt(0.5)), 'g', label='$\mu=-2, \ \sigma^2 = 0.5$')
#plt.xlim([-5,5])
plt.legend(loc='best')
plt.title("Log of Cumulative Distribtion Function \n Red curve is that standard normal distribution")
```

```python
plt.figure(figsize=(11,7))
x = numpy.linspace(-5, 5, 1000)
# loc:  mu or mean
# scale: sigma, scale, or std dev, sqrt(variance)
plt.plot(x, norm.sf(x, loc=0, scale = numpy.sqrt(0.2)), 'b', label='$\mu=0, \ \sigma^2 = 0.2$')
plt.plot(x, norm.sf(x, loc=0, scale = numpy.sqrt(1.0)), 'r', label='$\mu=0, \ \sigma^2 = 1.0$')
plt.plot(x, norm.sf(x, loc=0, scale = numpy.sqrt(5.0)), 'y', label='$\mu=0, \ \sigma^2 = 5.0$')
plt.plot(x, norm.sf(x, loc=-2, scale = numpy.sqrt(0.5)), 'g', label='$\mu=-2, \ \sigma^2 = 0.5$')
plt.xlim([-5,5])
plt.legend(loc='best')
plt.title("Survival Function \n Red curve is that standard normal distribution")
```

## Another way to draw random samples from a normal (Gaussian) Distribution

```python
def normpdf(xdata, mu, sigma):
    return 1./(numpy.sqrt(2*numpy.pi)*sigma)*numpy.exp(-0.5 * (1./sigma*(xdata - mu))**2)
```

```python
mu, sigma = 0, 1.0 # mean and standard deviation
num = 1000

s = numpy.random.normal(mu, sigma, num)

plt.figure(figsize=(11,7))


count, bins, ignored = plt.hist(s, 30, normed=True, alpha = 0.2)
x = numpy.linspace(min(bins), max(bins), 1000)
y = normpdf(x, mu, sigma)
plt.plot(x, y, 'b-', lw=4)

plt.plot(s, normpdf(s, mu, sigma), 'r.', linewidth=0)



plt.show()
```
